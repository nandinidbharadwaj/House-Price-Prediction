```python
import pandas as pd
house_filepath="C:/Users/Nandini D/Downloads/house_dataset.csv"
house_data=pd.read_csv(house_filepath)
filtered_house_data=house_data.dropna(axis=0)
y=filtered_house_data.Price
melbourne_features=['Area','Bedrooms','Age','Price']
X=filtered_house_data[melbourne_features]

from sklearn.tree import DecisionTreeRegressor
house_model=DecisionTreeRegressor()
house_model.fit(X,y)
```

```python
from sklearn.metrics import mean_absolute_error
predicted_house_prices=house_model.predict(X)
mean_absolute_error(y,predicted_house_prices)
```

```python
from sklearn.model_selection import train_test_split
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=0)
house_model=DecisionTreeRegressor()
house_model.fit(train_X,train_y)
value_predictions=house_model.predict(val_X)
print(mean_absolute_error(val_y,value_predictions))
```
